## Dados

### Gráfico compostos

```python
cod_compostos = dados['composto'].value_counts().index[0:5]  #.index: imprime somente os nomes, ñ a quantidade
#Entre colchetes informa qual o intervalo selecionado, portanto também pode ser alterada a ordem dos elementos(crescente e decrescente)
cod_compostos
```

```python
import seaborn as sns  #Biblioteca de visualização de dados
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(8, 6))
data = dados.query('composto in @cod_compostos')  #Dados do gráfico
ax = sns.countplot(x = 'composto' , data=data , order=cod_compostos)  #O order reorganiza os dados do query
#Contabiliza e seleciona os tipos de elementos de determinada coluna utilizando uma query
ax.set_title('Compostos mais relevantes', fontsize=18)  #Título do gráfico
ax.set_xlabel('Compostos', fontsize=14 , color='purple')
ax.set_ylabel('Quantidade', fontsize=14 , color='purple')
plt.show()  #Para não imprimir outras informações na saída
```

### Histogramas

```python
import matplotlib.pyplot as plt  #Aparência do gráfico
plt.title('Gene g0')
dados['g0'].hist(bins=50)  #Comando bins estipula o intervalo desejado das barras
#O valor atribuído entre parênteses, para o bins, é referente a quantidade de barras
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(8,6))
ax = sns.histplot(data=dados, x='g0', bins=100, hue='tratamento')
#                                                |Plota uma nova coluna da tabela
ax.set_title('g0 para diferentes tratamentos' , fontsize=18)
ax.set_xlabel('Expressão gênica' , fontsize=14 , color='purple')
ax.set_ylabel('Quantidade de id' , fontsize=14 , color='purple')

plt.show()
```

```python
dados.loc[:, 'g0':'g771'].describe().T  #Seleção mais ampla, de todos as colunas g
#T de transposição, converte linhas e colunas
```

```python
import seaborn as sns
import matplotlib.pyplot as plt  #Aparência do gráfico
sns.set()
plt.figure(figsize=(8,6))
ax = sns.histplot(data=dados.loc[:, 'g0':'g771'].describe().T['mean'] , bins=100)  #mean, indica qual a coluna selecionada
ax.set_title('Média-Genes' , fontsize=18)
ax.set_xlabel('Média dos genes' , fontsize=14 , color='purple')
ax.set_ylabel('Quantidade' , fontsize=14 , color='purple')
plt.show()
```

```python
import matplotlib.pyplot as plt  #Aparência do gráfico
plt.title('Máximo-Genes')
dados.loc[:, 'g0':'g771'].describe().T['max'].hist(bins=50)
plt.show()
```

```python
import matplotlib.pyplot as plt  #Aparência do gráfico
plt.title('Mínimo-Genes')
dados.loc[:, 'g0':'g771'].describe().T['min'].hist(bins=50)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt  #Aparência do gráfico
sns.set()
ax = sns.histplot(data=dados.loc[:, 'c0':'c99'].describe().T['mean'] , bins=100)  #mean, indica qual a coluna selecionada
ax.set_title('Média - Tipos celulares(c)' , fontsize=18)
ax.set_xlabel('Tipos celulares' , fontsize=14 , color='purple')
ax.set_ylabel('Quantidade' , fontsize=14 , color='purple')
plt.show()
```

### Boxplots

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.title('g0-geral')
sns.boxplot(x='g0' , data=dados)  #Utilizado para analisar visualmente a média dos dados selecionados
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.title('Tratamento gene g0')
sns.boxplot(y='g0' , x='tratamento' , data=dados)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.title('Tempo de exposição (g0)')
sns.boxplot(y='g0' , x='tempo' , data=dados)
plt.show()
```
